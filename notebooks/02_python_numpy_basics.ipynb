# Python and NumPy Basics

In this notebook, I will review the Python and NumPy concepts that I need for my stock price prediction project.

## Topics

- Python lists
- Loops
- Functions
- NumPy arrays
- Shape and data type
- Basic mathematical operations

```python
stock_prices = [100, 102, 101, 105, 108]

print("Stock prices:", stock_prices)
print("First price:", stock_prices[0])
print("Last price:", stock_prices[-1])
print("Number of prices:", len(stock_prices))
```

## My Notes

A Python list can store multiple values.

Index `0` gives the first value and index `-1` gives the last value.

The `len()` function returns the number of values in the list.

```python
stock_prices = [100, 102, 101, 105, 108]

for price in stock_prices:
    print("Price:", price)
```

## My Notes

A loop is used to repeat an operation for every value in a collection.

In this example, the loop processes each stock price one by one.

```python
def calculate_price_change(old_price, new_price):
    change = new_price - old_price
    return change


result = calculate_price_change(100, 108)

print("Price change:", result)
```

## My Notes

A function is a reusable block of code.

It can receive input values, perform an operation, and return a result.

Functions will help me organize repeated operations in the stock price project.

```python
import numpy as np

stock_prices = np.array([100, 102, 101, 105, 108])

print("NumPy array:", stock_prices)
print("Shape:", stock_prices.shape)
print("Data type:", stock_prices.dtype)
```

```python
print("Mean price:", np.mean(stock_prices))
print("Minimum price:", np.min(stock_prices))
print("Maximum price:", np.max(stock_prices))
print("Standard deviation:", np.std(stock_prices))
```

## My Notes

NumPy arrays are useful for numerical operations.

NumPy can quickly calculate values such as mean, minimum, maximum, and standard deviation.

These operations will help me analyze stock price data.

```python
stock_prices = np.array([100, 102, 101, 105, 108])

increased_prices = stock_prices + 5

print("Original prices:", stock_prices)
print("Increased prices:", increased_prices)
```

```python
stock_data = np.array([
    [100, 1000],
    [102, 1200],
    [101, 900],
    [105, 1500]
])

print(stock_data)
print("Shape:", stock_data.shape)
```
