# Pandas Basics

These skills will help me work with historical stock price data.

## Topics

- Series and DataFrame
- Creating a DataFrame
- Viewing rows
- Selecting columns
- Filtering data
- Basic statistics
- Missing values

```python
import pandas as pd

print("Pandas version:", pd.__version__)
```

```python
stock_data = {
    "Date": [
        "2026-07-01",
        "2026-07-02",
        "2026-07-03",
        "2026-07-04",
        "2026-07-05"
    ],
    "Close": [100, 102, 101, 105, 108],
    "Volume": [1000, 1200, 900, 1500, 1700]
}

df = pd.DataFrame(stock_data)

df
```

## My Notes

A DataFrame is a table made of rows and columns.

Each column represents a variable, and each row represents one observation.

In this example, the columns are Date, Close, and Volume.

```python
print("First rows:")
display(df.head())

print("Last rows:")
display(df.tail())

print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
```

```python
close_prices = df["Close"]

print(close_prices)
```

```python
high_price_days = df[df["Close"] > 102]

high_price_days
```

## My Notes

Filtering allows me to select rows that satisfy a condition.

For example, I can select days where the closing price or trading volume is above a certain value.

```python
print("Mean closing price:", df["Close"].mean())
print("Minimum closing price:", df["Close"].min())
print("Maximum closing price:", df["Close"].max())
print("Mean volume:", df["Volume"].mean())
```

```python
df.describe()
```

```python
df["Price_Change"] = df["Close"].diff()

df
```

```python
print(df.isnull())
```

## My Notes

NaN means that a value is missing or cannot be calculated.

The first price change is NaN because there is no previous day to compare it with.

Missing values must be checked before using data in a machine learning model.

```python
df["Date"] = pd.to_datetime(df["Date"])

print(df.dtypes)
```
