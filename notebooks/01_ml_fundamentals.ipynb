## 1. Dataset, Feature, and Target

A dataset is a collection of data used by a machine learning model.

- **Feature:** The input information used by the model.
- **Target:** The value that the model tries to predict.
- **Sample:** One row in the dataset.

Example:

| Study Hours | Exam Score |
|---|---:|
| 1 | 45 |
| 2 | 55 |
| 3 | 65 |
| 4 | 75 |

In this example:

- Study Hours is the feature.
- Exam Score is the target.
- Each row is one sample.

In my stock price project, past stock prices will be used to predict the next stock price.

## 2. Supervised and Unsupervised Learning

### Supervised Learning

Supervised learning means that the model learns from data where the correct answers are already known.

The dataset includes both features and targets.

My stock price project uses supervised learning because the model learns from past prices and their next values.

### Unsupervised Learning

Unsupervised learning means that the data does not include a target value.

The model tries to find patterns, groups, or similarities in the data by itself.

## 3. Regression and Classification

### Regression

Regression is used when the model predicts a numerical value.

Examples:

- Predicting a house price
- Predicting an exam score
- Predicting the next stock price

My stock price project is a regression problem because the output is a numerical price.

### Classification

Classification is used when the model predicts a category or class.

Examples:

- Spam or not spam
- Cat or dog
- Fraud or not fraud

## 4. Model, Training, and Prediction

### Model

A model is a system that learns patterns from data.

It takes features as input and produces a prediction as output.

### Training

Training is the process where the model learns from examples.

During training, the model compares its predictions with the correct target values and tries to reduce its mistakes.

### Prediction

Prediction is the result produced by the trained model for new data.

In my stock project, the model will learn from past prices and predict the next price.

## 5. Train Set and Test Set

### Train Set

The train set is the part of the dataset used to teach the model.

### Test Set

The test set is used to check how well the trained model works on unseen data.

We separate the data because testing the model on the same data used for training would not give a fair result.

In time series projects:

- Older data is used for training.
- Newer data is used for testing.

## 6. Overfitting and Data Leakage

### Overfitting

Overfitting happens when the model memorizes the training data instead of learning a general pattern.

The model may perform well on training data but poorly on new data.

### Data Leakage

Data leakage happens when the model sees information that it should not see during training.

In stock price prediction, future data must not be used to predict past data.

## 7. MSE and RMSE

### MSE

MSE stands for Mean Squared Error.

It measures the average squared difference between actual values and predicted values.

A lower MSE usually means better predictions.

### RMSE

RMSE stands for Root Mean Squared Error.

It is the square root of MSE.

RMSE is easier to understand because it uses the same unit as the target value.

```python
import numpy as np
from sklearn.metrics import mean_squared_error

actual_values = [70, 80, 90]
predicted_values = [68, 84, 88]

mse = mean_squared_error(actual_values, predicted_values)
rmse = np.sqrt(mse)

print("MSE:", mse)
print("RMSE:", rmse)
```
