## 7. MSE and RMSE
### MSE

MSE stands for Mean Squared Error.

It measures the average squared difference between the actual values and the predicted values.

A lower MSE usually means that the model's predictions are closer to the real values.
### RMSE

RMSE stands for Root Mean Squared Error.

It is the square root of MSE.

RMSE is easier to understand because it is expressed in the same unit as the target value.
### Example

Actual exam scores:

- 70
- 80
- 90

Predicted exam scores:

- 68
- 84
- 88

The errors are:

- 70 - 68 = 2
- 80 - 84 = -4
- 90 - 88 = 2


## 6. Overfitting and Data Leakage
### Overfitting

Overfitting happens when a model learns the training data too closely.

The model may perform very well on the training data but poorly on new test data.

This means the model memorized the training examples instead of learning a general pattern.
### Data Leakage

Data leakage happens when information from the test data is accidentally used during training.

This gives the model unfair information and makes the results look better than they really are.

In time series projects, future data must not be used to predict the past.


Overfitting means the model memorizes the training data and cannot work well on new data.

Data leakage means the model sees information that it should not see during training.

In my stock project, older prices will be used for training and newer prices will be used for testing.


## 5. Train Set and Test Set
### Train Set

The train set is the part of the dataset used to teach the model.

The model sees both the features and the correct target values in this part.

It learns the relationship between input and output from the train set.
### Test Set

The test set is used to check how well the trained model works on new data.

The model should not see the test data during training.

This helps us understand whether the model really learned the pattern or only memorized the training data.


The train set is used to teach the model.

The test set is used to check the model with unseen data.

We separate the data because testing the model on the same data used for training would not give a fair result.
